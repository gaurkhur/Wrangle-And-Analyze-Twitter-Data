# INSIGHTS GATHERED FROM TWITTER ARCHIVES FOR @dog_rates



## Insight 1
### Software platform chosen for tweets

Of the 2278 cleaned tweets that are present in the merged dataset (df_twitter), most of the tweets have been posted/rated from twitter app on iphone in the years 2015, 2016, and 2017. 




## Insight 2 
### Breeds with highest and lowest average ratings

Of the 113 different breeds in the dataset, Bouvier Des Flandres, Saluki, Briard and Tibetian Mastiff have the highest ratings; while Scotch Terrier, Walker Hound, Soft-Coated Wheaten Terrier, Japanese Spaniel breeds have the lowest ratings.

## Insight 3 
### Breeds with highest favorite and retweet count
